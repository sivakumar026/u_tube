```python
red.dropna(inplace=True)
twi.dropna(inplace=True)

red = red.drop_duplicates()
twi = twi.drop_duplicates()
```

```python
red = red[["clean_comment", "category"]]
twi = twi[["clean_text", "category"]]
```

```python
red = red.rename(columns={"clean_comment":"text"})
twi = twi.rename(columns={"clean_text":"text"})
```

```python
df = pd.concat([red, twi], ignore_index=True)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z ]", "", text)
    return text

df["text"] = df["text"].apply(clean_text)
```
